```python
def convert_from_csv_to_json(df, csv_file_name):
    df.to_csv(csv_file_name, index = False)
    # Open a csv reader called DictReader
    with open(csv_file_name, encoding = 'utf-8') as csvf:
        csv_reader = csv.DictReader(csvf)
        csv_header = df.columns
        print(csv_reader)
        json_list = []
        for rec in csv_reader:
            #print(rec)
            json_list.append(rec)
    
    return json_list
```

# Creating order line collection in mongodb

```python
query = """
SELECT     ord_dtl.*
         , dtl_agg.total_products_purchased
         , ord_hdr.customer_id
         , cust_agg.total_orders
         , cust_tot_ord_qty.total_purchased
FROM       order_details ord_dtl
INNER JOIN orders        ord_hdr
ON         ord_hdr.order_id = ord_dtl.order_id
INNER JOIN customers cust
ON         ord_hdr.customer_id = cust.customer_id
INNER JOIN (SELECT customer_id
                 , COUNT(order_id) AS total_orders 
            FROM   orders
            GROUP BY customer_id
           ) cust_agg
ON         cust_agg.customer_id = ord_hdr.customer_id
INNER JOIN (SELECT     cust.customer_id
                     , dtl.product_id
                     , SUM(dtl.quantity) AS total_products_purchased
            FROM       customers     cust
            INNER JOIN orders        hdr
            ON         cust.customer_id = hdr.customer_id
            INNER JOIN order_details dtl
            ON         hdr.order_id = dtl.order_id
            GROUP BY   cust.customer_id
                     , dtl.product_id
           ) dtl_agg
ON         dtl_agg.customer_id = cust.customer_id
AND        dtl_agg.product_id  = ord_dtl.product_id
INNER JOIN (SELECT     cust.customer_id
                     , SUM(dtl.quantity) AS total_purchased
            FROM       customers     cust
            INNER JOIN orders        hdr
            ON         cust.customer_id = hdr.customer_id
            INNER JOIN order_details dtl
            ON         hdr.order_id = dtl.order_id
            GROUP BY   cust.customer_id
           ) cust_tot_ord_qty
ON         cust_tot_ord_qty.customer_id = cust.customer_id
;
"""
order_line_df = pg_helper.select_query_pandas(query)
order_line_df['rating'] = order_line_df['total_products_purchased'] / order_line_df['total_purchased']
f_csv_file = "order_line.csv"
json_list = convert_from_csv_to_json(df = order_line_df, csv_file_name = f_csv_file)
#order_line_collection.drop()
order_line_collection = mongo_db["order_line_collection"]
inserted_order_line_documents = order_line_collection.insert_many(json_list)
print(f"Total order_lines : {len(order_line_df)}")
print(inserted_order_line_documents.inserted_ids)
```

# Customer Point of View in MongoDB

```python
query = """
SELECT     cust.customer_id
         , cust.company_name                         AS customer_name
         , cust.country                              AS customer_country
         , COUNT(ord_hdr.order_id)                   AS total_orders
         , dtl_agg.total_quantity_purchased          AS total_quantity_purchased
         , dtl_agg.total_spend_amount                AS total_spend_amount
         , dtl_agg.total_unique_products_purchased   AS total_unique_products_purchased

FROM       customers     cust
INNER JOIN orders        ord_hdr
ON         ord_hdr.customer_id = cust.customer_id
INNER JOIN (SELECT     cust.customer_id
                     , SUM(dtl.quantity)                                            AS total_quantity_purchased
                     , SUM((dtl.unit_price - dtl.discount) * dtl.quantity)          AS total_spend_amount
                     , COUNT(DISTINCT dtl.product_id)                               AS total_unique_products_purchased
            FROM       customers     cust
            INNER JOIN orders        hdr
            ON         cust.customer_id = hdr.customer_id
            INNER JOIN order_details dtl
            ON         hdr.order_id = dtl.order_id
            GROUP BY   cust.customer_id
           ) dtl_agg
ON         dtl_agg.customer_id = cust.customer_id

GROUP BY   cust.customer_id
         , cust.company_name
         , cust.country
         , dtl_agg.total_quantity_purchased          
         , dtl_agg.total_spend_amount
         , dtl_agg.total_unique_products_purchased
         
ORDER BY   dtl_agg.total_spend_amount  DESC    
;
"""
cust_stats_df = pg_helper.select_query_pandas(query)
pd.set_option('precision', 2)
print(cust_stats_df.head())
cust_stats_df = pg_helper.select_query_pandas(query)
f_csv_file = "cust_stats.csv"
json_list = convert_from_csv_to_json(df = cust_stats_df, csv_file_name = f_csv_file)
#cust_stats_collection.drop()
cust_stats_collection = mongo_db["cust_stats_collection"]
inserted_cust_stats_documents = cust_stats_collection.insert_many(json_list)
print(f"Total cust_statss : {len(cust_stats_df)}")
print(inserted_cust_stats_documents.inserted_ids)
```

# Region Point of View in MongoDB

```python
query = """

SELECT     agg.country                                AS country
         , SUM(agg.total_quantity_purchased)          AS country_total_orders
         , SUM(agg.total_quantity_purchased)          AS country_total_quantity_purchased
         , SUM(agg.total_spend_amount)                AS country_total_spend_amount
         , SUM(agg.total_unique_products_purchased)   AS country_total_unique_products_purchased
FROM
(
SELECT     cust.customer_id
         , cust.company_name                         AS customer_name
         , cust.country                              AS country
         , COUNT(ord_hdr.order_id)                   AS total_orders
         , dtl_agg.total_quantity_purchased          AS total_quantity_purchased
         , dtl_agg.total_spend_amount                AS total_spend_amount
         , dtl_agg.total_unique_products_purchased   AS total_unique_products_purchased

FROM       customers     cust
INNER JOIN orders        ord_hdr
ON         ord_hdr.customer_id = cust.customer_id
INNER JOIN (SELECT     cust.customer_id
                     , SUM(dtl.quantity)                                            AS total_quantity_purchased
                     , SUM((dtl.unit_price - dtl.discount) * dtl.quantity)          AS total_spend_amount
                     , COUNT(DISTINCT dtl.product_id)                               AS total_unique_products_purchased
            FROM       customers     cust
            INNER JOIN orders        hdr
            ON         cust.customer_id = hdr.customer_id
            INNER JOIN order_details dtl
            ON         hdr.order_id = dtl.order_id
            GROUP BY   cust.customer_id
           ) dtl_agg
ON         dtl_agg.customer_id = cust.customer_id

GROUP BY   cust.customer_id
         , cust.company_name
         , cust.country
         , dtl_agg.total_quantity_purchased          
         , dtl_agg.total_spend_amount
         , dtl_agg.total_unique_products_purchased
         
) agg

GROUP BY   agg.country

ORDER BY   SUM(agg.total_spend_amount)  DESC            
;
"""
region_stats_df = pg_helper.select_query_pandas(query)
f_csv_file = "region_stats.csv"
json_list = convert_from_csv_to_json(df = region_stats_df, csv_file_name = f_csv_file)
#region_stats_collection.drop()
region_stats_collection = mongo_db["region_stats_collection"]
inserted_region_stats_documents = region_stats_collection.insert_many(json_list)
print(f"Total region_statss : {len(region_stats_df)}")
print(inserted_region_stats_documents.inserted_ids)
```

# Product Point of View in MongoDB

```python
query = """
SELECT     agg.product_id
         , agg.product_name
         , SUM(agg.total_quantity_purchased) AS total_quantity_purchased
         , SUM(agg.total_spend_amount)       AS total_spend_amount
         
FROM
(
SELECT     dtl_agg.product_id                        AS product_id
         , dtl_agg.product_name                      AS product_name
         , dtl_agg.total_quantity_purchased          AS total_quantity_purchased
         , dtl_agg.total_spend_amount                AS total_spend_amount

FROM       customers     cust
INNER JOIN orders        ord_hdr
ON         ord_hdr.customer_id = cust.customer_id
INNER JOIN (SELECT     cust.customer_id
                     , dtl.product_id
                     , prod.product_name
                     , SUM(dtl.quantity)                                            AS total_quantity_purchased
                     , SUM((dtl.unit_price - dtl.discount) * dtl.quantity)          AS total_spend_amount
            FROM       customers     cust
            INNER JOIN orders        hdr
            ON         cust.customer_id = hdr.customer_id
            INNER JOIN order_details dtl
            ON         hdr.order_id = dtl.order_id
            INNER JOIN products      prod
            ON         dtl.product_id = prod.product_id
            WHERE      prod.discontinued = 0
            GROUP BY   cust.customer_id
                     , dtl.product_id
                     , prod.product_name
           ) dtl_agg
ON         dtl_agg.customer_id = cust.customer_id

GROUP BY   dtl_agg.product_id
         , dtl_agg.product_name
         , dtl_agg.total_quantity_purchased          
         , dtl_agg.total_spend_amount
         
ORDER BY   dtl_agg.total_spend_amount  DESC 
) agg
GROUP BY agg.product_id
       , agg.product_name
ORDER BY SUM(total_spend_amount) DESC
;
"""
product_stats_df = pg_helper.select_query_pandas(query)
pd.set_option('precision', 2)
print(product_stats_df.head())
product_stats_df = pg_helper.select_query_pandas(query)
f_csv_file = "product_stats.csv"
json_list = convert_from_csv_to_json(df = product_stats_df, csv_file_name = f_csv_file)
#product_stats_collection.drop()
product_stats_collection = mongo_db["product_stats_collection"]
inserted_product_stats_documents = product_stats_collection.insert_many(json_list)
print(f"Total product_statss : {len(product_stats_df)}")
print(inserted_product_stats_documents.inserted_ids)
```
